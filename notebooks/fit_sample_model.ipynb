```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from catboost import CatBoostClassifier
from metricsreport import MetricsReport
```

```python
VERSION = '2025-10-26'
```

# Data

```python
# Load breast cancer dataset
data = load_breast_cancer(as_frame=True)
```

```python
data.data
```

```python
data.data.columns
```

```python
# same fix feature name
data.data.columns = [s.replace(' ', '_') for s in data.data.columns]
```

```python
data.data.columns
```

```python
data.data.info()
```

# Split data

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.25, random_state=42)

print(X_train.shape, X_test.shape)
```

# Model

```python
# Create CatBoost classifier
model = CatBoostClassifier(verbose=False)
# Train the model on the training set
model = model.fit(X_train, y_train)
```

```python
import os

# Define the full path to the directory
MODEL_DIR = '../app/models/'
VERSION = '2025-10-26' # or whatever variable you are using for the version

# 1. Create the directory if it does not exist
os.makedirs(MODEL_DIR, exist_ok=True)

# 2. Save the model
model.save_model(f'{MODEL_DIR}catboost_model_{VERSION}.cbm')
```

```python
# Make predictions on the test set
y_pred = model.predict_proba(X_test)[:, 1]
```

# Metrics


```python
mr = MetricsReport(y_test, y_pred)
mr.print_metrics()
```

```python
mr.plot_roc_curve().show()
```
